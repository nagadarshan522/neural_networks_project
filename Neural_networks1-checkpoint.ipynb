```python
import tensorflow as tf

import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, roc_curve, auc

from keras.models import Sequential
from keras.layers import Dense

import matplotlib.pyplot as plt
```

```python
#loading the dataset
df=pd.read_csv("C:\\Users\\Lenovo\\Downloads\\gas_turbines.csv")
df.head()
```

# preprocessing

```python
df.info()
```

```python
df.describe()
```

```python
sns.pairplot(df)
```

```python
#Normalization
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler().fit(df.iloc[:,:])
df.iloc[:,:]=scaler.transform(df.iloc[:,:])
df
```

```python
#defining labelled data and features
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#building the model
perceptron_model = Sequential()

perceptron_model.add(Dense(1, input_dim=10, activation='sigmoid', kernel_initializer='normal'))
```

```python
perceptron_model.compile(loss='mse', optimizer='adam', metrics=['mae','mse'])
```

```python
perceptron_model.summary()
```

```python
train=perceptron_model.fit(x_train, y_train, epochs=30, batch_size=64)
```

```python
train
```

```python
test_pred=perceptron_model.predict(x_test)
train_pred=perceptron_model.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error
rmse=mean_squared_error(y_test,test_pred,squared=False)
rmse
```
